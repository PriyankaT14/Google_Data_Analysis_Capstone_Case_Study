### 5.1 Data Visualization and Insights

#### FitBit App Usage Vs Day_of_the_Week

```python
WeekDays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.set_style("whitegrid")
plt.figure(figsize = (10,6))
ax = sns.countplot(x ='Day_of_the_Week', data = daily_activities, order = WeekDays, palette = 'pastel')
ax.set(ylabel='Bar Count', title='FitBit App Usage Vs Day Of The Week')
# add annotations
for c in ax.containers:
    
    # custom label calculates percent and add an empty string so 0 value bars don't have a number
    labels = [f'{h/daily_activities.Day_of_the_Week.count()*100:0.1f}%' if (h := v.get_height()) > 0 else '' for v in c]
    
    ax.bar_label(c, labels=labels, label_type='edge')
plt.xticks(rotation=90)
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Week Days when app was used most: Tuesday, Wednesday, Thursday

There was decline in the usage of app on: Monday, Friday, Saturday and Sunday

#### Calories Vs TotalSteps

```python
# Calories Vs TotalSteps
plt.figure(figsize = (15,6))
plt.scatter (x = 'TotalSteps', y = 'Calories', data = daily_activities, alpha = 0.6, cmap = 'spring', c = daily_activities.Calories)
mean_steps = round(daily_activities.TotalSteps.mean())
mean_calories = round(daily_activities.Calories.mean())
plt.axvline(mean_steps, color = 'g', label = 'Average_Steps')
plt.axhline(mean_calories, color = 'r', label = 'Average_Calories_Burned')
plt.title('Calories Burned Vs TotalSteps')
plt.xlabel('TotalSteps')
plt.ylabel('Calories')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

From above plot:

1. Calories and TotalSteps have positive relation

2. TotalSteps > 10000 more calories are burned
    

#### Calories Vs Total_Hours

```python
# Calories Vs Total_Hours
plt.style.use('default')
plt.figure(figsize = (15,6))
plt.scatter (x = 'Calories', y = 'Total_Hours', data = daily_activities, alpha = 0.6, cmap = 'spring', c = daily_activities.Calories)
mean_hours = round(daily_activities.Total_Hours.mean())
mean_calories = round(daily_activities.Calories.mean())
plt.axvline(mean_hours, color = 'g', label = 'Average_Hours')
plt.axhline(mean_calories, color = 'r', label = 'Average_Calories_Burned')
plt.title('Total_Hours Vs Calories')
plt.xlabel('Calories')
plt.ylabel('Total_Hours')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

From above plot:

1. Relation between Total_Hours and Calories is weak

#### Percentage Activity in Minutes

```python
# Total individual minutes
plt.style.use('default')
VeryActiveMinutes_Sum = daily_activities['VeryActiveMinutes'].sum()
FairlyActiveMinutes_Sum = daily_activities['FairlyActiveMinutes'].sum()
LightlyActiveMinutes_Sum = daily_activities['LightlyActiveMinutes'].sum()
SedentaryMinutes_Sum = daily_activities['SedentaryMinutes'].sum()
# Plot
slices = [VeryActiveMinutes_Sum, FairlyActiveMinutes_Sum, LightlyActiveMinutes_Sum, SedentaryMinutes_Sum]
labels = ['Very_Active_Minutes', 'Fairly_Active_Minutes', 'Lightly_Active_Minutes', 'Sedentary_Minutes']
colours = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']
explode = [0, 0, 0, 0.1]
plt.pie(slices, labels = labels, colors = colours, wedgeprops = {"edgecolor": "black"}, explode = explode, autopct = "%1.1f%%")
plt.title('Percentage Activity in Minutes')
plt.show()
```

Large portion of the users on FitBit are mostly:
1. Sedentary Users(81.3%)
2. Lightly Active User(15.8%) 

FitBit App is used less for track fittness/running as: Fairly Active Users are only 1.1% and Very Active User are 1.7% only.
